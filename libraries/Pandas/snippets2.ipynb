# Add or remove rows

```python
import pandas as pd
```

```python
# dictionary 
cars = {
    "brand": ["Ford", "BMW", "Toyota", "Honda", "Ford", "BMW", "Toyota", "Honda", "Ford", "BMW", "Toyota", "Honda", "Ford", "BMW", "Toyota", "Honda"],
    "electric": [False, True, False, True, False, True, False, True, False, True, False, True, False, True, False, True],
    "year": [1964, 2022, 2019, 2021, 2018, 2021, 2020, 2023, 2017, 2022, 2023, 2021, 2019, 2021, 2018, 2022],
    "engine": ['F5', 'B7', 'T6', 'H4', 'F4', 'B6', 'T5', 'H3', 'F9', 'B8', 'T4', 'H2', 'F6', 'B5', 'T3', 'H1']
}

df=pd.DataFrame(cars)
df
```

```python
df['model']= df['brand']+' '+df['engine']
df
```

```python
df['model'].str.split(' ',expand=True) # ' ' الحاجه الي هتفصل بينهم 
# expand --> seprate each in one col
```

```python
# df.append({'brand': 'BMW'}, ignore_index=True) deprecated !!!!!!!!!!!!!
```

```python
df.drop(columns=['model']) 
df.drop(index=3) # drop index means row !
```
