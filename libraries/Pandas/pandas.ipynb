# Cleaning the missing values

```python
import numpy as np


people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
peopledata=pd.DataFrame(people) # missing and NA is custom null
peopledata
```

```python
# null should be np.nan null(not a number) or None
peopledata.dropna(axis='columns') # drop the columns which have null  
```

```python
peopledata.dropna(axis='index',how='all') # all the row ( index )  is null
```

```python
peopledata.dropna(axis='index',how='any') # any value in the row ( index )  is null
```

```python
peopledata.dropna(axis='index',how='any',subset=['last'])
```

```python
# Keep only the rows with at least 1 non-NA values.
peopledata.dropna(thresh=1)
```

```python
peopledata.drop(columns=['first']) # drop specified column
```
